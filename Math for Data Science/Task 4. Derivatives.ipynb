```python
"""
Уровень 0:
Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.

"""
```

```python
from scipy.misc import derivative
import math

def f(x):
    return x ** 5 + 4 * math.sin(2 * x) + math.cos(3 * x + 3)

x = 1

print('f\'(', x, '):', derivative(f, x, n = 1, dx = 1e-6))
print('f\'\'(', x, '):', derivative(f, x, n = 2, dx = 1e-6))
```

```python
"""
Уровень 1:
Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].
"""
```

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return math.sin(2 * x + 1) ** 5

x = np.linspace(-5, 5)
f1x = [derivative(f, xx, dx = 1e-6, n = 1) for xx in x]
f2x = [derivative(f, xx, dx = 1e-6, n = 2) for xx in x]

plt.plot(x, f1x, label = '1я производная')
plt.plot(x, f2x, label = '2я производная')
plt.legend()
plt.show()
```
