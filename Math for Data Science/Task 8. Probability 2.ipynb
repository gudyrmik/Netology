```python
'''
Домашнее задание "Теория вероятности. Непрерывные Случайные Величины".
Уровень 0:
Задание 1:
Получить интуицию, на что влияют параметры распределений.
Построить по 5 функций плотностей с разными параметрами для каждого распределения:
нормальное, экспоненциальное, Стьюдента на одном графике.
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
x = np.linspace(0, 10, 1000)
mus = np.linspace(0, 10, 5)
sigmas = np.linspace(0.1, 1)

for i in range(len(sigmas)):
    nd = sts.norm(mus[i], sigmas[i])
    cdf = nd.cdf(x)
    
    plt.plot(x, cdf, label=f'{i}')

plt.title('Нормальная красотища')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.show()
```

```python
lams = np.linspace(0.1, 1, 5)

for l in range(len(lams)):
    ed = sts.expon(l)
    cdf = ed.cdf(x)
    
    plt.plot(x, cdf, label=f'{l}')

plt.title('Экспоненциальная красотища')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.show()
```

```python
ns = np.linspace(1, 10, 5)

for n in range(len(ns)):
    sd = sts.t(n)
    cdf = sd.cdf(x)
    
    plt.plot(x, cdf, label=f'{n}')

plt.title('Стьюдентская красотища')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.show()
```
