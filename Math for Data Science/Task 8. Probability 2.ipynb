```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
x = np.linspace(0, 10, 1000)
mus = np.linspace(0, 10)
sigmas = np.linspace(0.1, 1)

for i in range(len(sigmas)):
    nd = sts.norm(mus[i], sigmas[i])
    cdf = nd.cdf(x)
    
    plt.plot(x, cdf, label=f'{i}')

plt.title('Красотища')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.show()
```
